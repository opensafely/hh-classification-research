```python
%matplotlib inline
import numpy as np
import scipy.stats as st
from scipy import sparse
import matplotlib.pyplot as plt
import pandas as pd
from tqdm.notebook import tqdm
```

```python
df = pd.read_csv("~/Documents/Work/OpenSafely/households-research/release/20200623_release_edited.csv")
df
```

```python
hids = pd.unique(df.household_id)
```

```python
hhs = []
for hi, hid in tqdm(enumerate(hids)):
    dfh = df[df.household_id == hid]
    hhs.append(len(pd.unique(dfh.patient_id)))
hh_sizes = np.array(hhs)
```
